# K-nearest Neighbour Introduction
- simple machine learning Algorithm.
- used for both regression as well as classification.
- supervised learning algorithm
- No training is involved 0(1)
- All time is query time, we compute the distance of current point with all other point., since there are N points so complexity is 0(N)
- Non Parametric Algorithm.
-it is used as a baseline algorithm in many applications.
- Base line means- any algorithm you want to use will have accuracy greater than k-nn algorithm.
- it is basically a brute force algorithm in machine learning.

# How it works?
- suppose we have two classes - orange and apples and now there is a data point for which we have to predct whether it belongs to class apple or orange. so what  we will do is- here k is the no of nearer points we will take into consideration. now suppose we took 5 points for prediction. so k=5. now from 5 , 3 is from apple and 2 is from orange. so we will now do voting. so since 3 are from apple so the given data point will belong to apple class.

# Implementation

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=399,n_features=2,n_informative=2,n_redundant=0)
print(x.shape)
print(y.shape)
```

```python
print(y)
```

```python
plt.style.use('seaborn')
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
query_x=np.array([2,3])
plt.scatter(query_x[0],query_x[1],color='red')
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
print(query_x)
```

```python
# Now we want to find out whether this query point belongs to class 0 or class 1
```

```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
def KNN(x,y,query_point,k=5):
    val=[]
    m=x.shape[0]
    for i in range(m):
        d=distance(query_point,x[i])
        val.append((d,y[i]))
    val=sorted(val)
    val=val[:k]
    val=np.array(val)
    new_val=np.unique(val[:,1],return_counts=True)
    print(new_val)
    index=new_val[1].argmax()
    pred=new_val[0][index]
    return pred
    
```

```python
val=KNN(x,y,query_x)
print(int(val))
```
